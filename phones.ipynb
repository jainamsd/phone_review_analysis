```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\Jainam\Downloads\raw_data.csv")
```

```python
data
```

```python
#getting Information on dataframe
data.info()
```

```python
#checking for null values
data.isnull()
```

```python
#checking total number of of null values
data.isnull().sum()
```

```python
#replacing null values with mean of each column
for col in data.columns[3:8]:
    data[col]=data[col].replace(np.NAN,data[col].mean())
print(data)
```

```python
#checking duplicated entries 
data['model'].duplicated().sum()
```

```python
#typecasting price 
data['price']=data['price'].str.slice(1,).astype(int)
```

```python
#typecasting date
# data['launch']=data['launch'].astype(datetime64[ns])
```

```python
#updated data frame
data.info()
data
```

```python
#getting company name from model
data['company']=data['model'].str.partition(" ")[0]
data.info()
```

```python
#number of models per company
no_of_models=data.groupby('company').size().sort_values(ascending=False)
```

```python
#Visualizing company with most no.of models 
Most_models=no_of_models.head()
```

```python
Most_models.plot.bar(xlabel="Company",ylabel="No.of Models")
```

```python
#identifying correlation
correlation_matrix=data.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,annot=True)
##We can see the correlation between camera rating and price 
```

```python
##Top 5 expensive phones
top_five_rating = data.sort_values(by="price" ,ascending = False).head()

top_five_rating

# [model_name,top_price]=[data['model'],data['price'].sort_values(ascending=False).head()]
# [model_name,top_price]
```

```python
##bar plot for most expensive phones
gr=sns.barplot(x = 'model', y = 'price', data= top_five_rating)
gr.set_xticklabels(gr.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
